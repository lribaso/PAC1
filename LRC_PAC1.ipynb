### Pregunta 1 **<font color="blue" size="+2">EG</font>**
¿Què són els f-strings en Python i com s'utilitzen per a imprimir informació per pantalla? ¿Quines altres formes similars existeixen en Python?

**Posa un exemple de cadascun d'ells**.

**Resposta:**


Les **f-strings** (formatted strings) en Python són una manera d’incloure valors de variables o expressions dins d’una cadena de text. Es creen posant una f davant de la cadena i escrivint les variables o expressions dins dels claudàtors {}. De manera que, a l'hora d'imprimir la informació, la variable envoltada de {} sigui substituïda per el seu valor.  

També existeixen altres maneres similars per fer el mateix que les f-strings:

*   Una d'elles seria utilitzar l'expressió **format()**, que permet substituïr {} dins d'una cadena per altres variables.  
*   També es podria fer utilitzant **wildcards** (comodins) representats com a % amb una lletra (depenent de quin tipus de dada sigui) dins de la cadena.
*   O una altra manera seria la més rudimentària, anar **concatenant** diferents cadenes i variables.


> *Per respondre aquesta pregunta s’ha consultat la documentació oficial de Python sobre el mòdul string (Python Software Foundation): https://docs.python.org/3/library/string.html.*


````python
# Primer posarem un exemple de les f-strings
nom = "Laura"
edat = 20
print(f"Hola em dic {nom} i tinc {edat} anys.")

# Ara fem el mateix però fent servir l'expressió format()
print("Hola em dic {} i tinc {} anys.".format(nom, edat))

# Utilitzant comodins
print("Hola em dic %s i tinc %d anys." % (nom, edat))

# També es pot fer concatenant cadenes i les variables
print("Hola em dic " + nom + " i tinc " + str(edat) + " anys.")
````

Python disposa d'un *idiom* molt útil conegut com list comprehensions. Pots explicar en què consisteix? Indiqueu
**dos exemples** de codi que utilitzin aquest *idiom* i documenteu cada part de codi rellevant.

**Resposta:**

L'*idiom* conegut com a **list comprehensions** es refereix una altra manera de definir llistes, a partir d'altres llistes ja definides. És a dir, es poden crearnoves llistes descrivint propietats compartides pels seus membres (comprehensions), en comptes de definir la llista a partir d'enumeració.

### Exercici 7 **<font color="blue" size="+2">EG</font>**


Escriu un programa que calculi el nombre de dies entre aquestes dues dates:

`01/12/1964`

`31/12/2025`

El resultat s'ha d'assignar a la variable `dies` i s'ha d'imprimir per pantalla Aixi doncs, el resultat per pantalla del'execució del teu programa deu ser:
 ```
 22310
 ```


````python
# Aquí per simplificar ja importem només la classe datetime directament
from datetime import datetime

# Convertim les dates escrites com a cadenes de text (strings) en objectes datetime utilitzant strptime().
data_inici = datetime.strptime("01/12/1964", "%d/%m/%Y")
data_fi = datetime.strptime("31/12/2025", "%d/%m/%Y")

# Restem les dues dates per obtenir un objecte timedelta (diferència de temps).
# Amb l'atribut .days obtenim el nombre total de dies com a enter.
dies = (data_fi - data_inici).days
print(dies)
````
